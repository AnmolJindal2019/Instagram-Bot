```python
def searchResults(text):
    searchbox = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Search']")))
    searchbox.clear()
    searchbox.send_keys(text)
    time.sleep(5)
    search_results = driver.find_elements_by_xpath("//div[contains(@class , 'fuqBx')]//a")
    results = []
    for i in search_results:
        s = i.get_attribute('href')
        s = s.split('/')
        if(len(s)==5):
            results.append(s[3])
    return results
```

```python
#Question 2.1
result = searchResults('food')
for i in range(1 , len(result)+1):
    print(i,'. ',result[i-1])
```

```python
def openProfile(profile):
    searchbox = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH, "//input[@placeholder='Search']")))
    searchbox.clear()
    time.sleep(3)
    searchbox.send_keys(profile)
    time.sleep(3)
    button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[contains(@class, "fuqBx")]/div/a')))
    button.click()
    
```

```python
def firstPost():
    post = driver.find_element_by_class_name("kIKUG");
    post.click()
```

```python
from bs4 import BeautifulSoup as bs
def likePost():
    time.sleep(3)
    heart = driver.find_element_by_class_name("fr66n")
    data = bs(heart.get_attribute('innerHTML'),'html.parser')
    if(data.find('svg')['aria-label'] == 'Like'):
        heart.click()
        print("Liked Now.")
    else:
        print("Already Liked.")
    time.sleep(3)
    
```

```python
def nextPost():
    time.sleep(2)
    try:
        arrow = driver.find_element_by_class_name("coreSpriteRightPaginationArrow")
        time.sleep(2)
        return arrow
    except selenium.common.exceptions.NoSuchElementException:
        return 0

def likeRange(n):
    i=2
    while(True):
        if i==n+1:
            break
            
        np = nextPost()
        if np != False:
 
            np.click()
            time.sleep(2)
         
            print("Post No." ,i ,": ",end = "")
            likePost()
            time.sleep(2)
        else:
            print("No more post left.")
            break
        i+=1
```

```python
def scroll():
    for i in range(50):
        current_height= driver.execute_script('return document.getElementsByClassName("isgrP")[0].scrollHeight;')
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTo(0, arguments[0]);',current_height) 
        time.sleep(2)
```

```python
def getFollowers(n):
    i=0
    titles = driver.execute_script('return document.getElementsByClassName("FPmhX");')
    li_followers = []
    for ele in titles:
        li_followers.append(ele.text)
        i += 1
        if i == n:
            break
    btnClose = driver.find_elements_by_class_name('WaOAr')[-1].find_element_by_xpath('button[contains(@class, "wpO6b")]').click()
    return li_followers
```

```python
def extractFollowers(profile , num):
    openProfile(profile)
    time.sleep(2)
    driver.find_element_by_partial_link_text("follower").click()
    time.sleep(2)
    scroll()
    time.sleep(2)
    followers = getFollowers(num)
    return followers
    
```

```python
def getFollowing(profile):
    openProfile(profile)
    time.sleep(2)
    driver.find_element_by_partial_link_text("Followed by").click()
    time.sleep(2)
    li = driver.find_element_by_class_name('PZuss')
    li1 = li.find_elements_by_xpath('//span[contains(@class, "Jv7Aj mArmR MqpiF  ")]/a')
    followers = []
    for i in range(1,3):
        ele = li.find_element_by_xpath('/html/body/div[5]/div/div/div[2]/ul/div/li['+str(i)+']/div/div[3]/button')
        if ele.text == 'Following':
            continue
        followers.append(li1[i].get_attribute('href'))
    return followers
```

```python
#Question 6.2
followers = getFollowing("foodtalkindia")
if len(followers) == 0:
    print('All Following back.')
else:
    for i in followers:
        i = i.split('/')
        print(i[3])
```
